```python
lemma = open("lemma.al.txt", "r")
count = 0
word_list = []
for line in lemma.readlines():
    words = line.split(" ")
    if words[2].isalpha():
        word_list.append(words[2])

print(all(True if word.isalpha() else False for word in word_list))
```

```python
dfBrands = pd.DataFrame.from_csv("brand_data_singleword.csv")
dfOtherBrands = dfBrands.iloc[0:0]
count = 0
for row in dfBrands.iterrows():
    if row[1][0] in word_list:
        dfOtherBrands = dfOtherBrands.append(list(row[1:]))
        dfBrands.drop(row[0], inplace=True)  
        count += 1
count
```

```python
dfOtherBrands.to_csv("english_word_brands.csv") #66 brands
dfBrands.to_csv("non_english_word_brands.csv") #1021 brands
```
